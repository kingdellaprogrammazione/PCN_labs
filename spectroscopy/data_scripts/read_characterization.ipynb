```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
base = Path.cwd().parent
print(base)
```

```python
data_file = base / 'data' / 'characterization' / 'charact.csv'
```

```python
df = pd.read_csv(data_file)
print(df)
```

```python
df = df.dropna(axis=1, how='all')
df = df.dropna(axis=0, how='all')
```

```python
print(df)
```

```python
X = []
Y = []

for row in df.values:
    if row[2] == 'mW':
        y = row[1] * 1e3
    else:
        y = row[1]
    x = row[0]
    X.append(x)
    Y.append(y)
```

```python
plt.plot(X, Y, '.')
#plt.xlim(50,65)
#plt.ylim(0,128)
plt.grid()
```
